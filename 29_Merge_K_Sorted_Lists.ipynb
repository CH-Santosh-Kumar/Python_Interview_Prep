# Problem 29 : Merge K Sorted Lists
### You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.
### Merge all the linked-lists into one sorted linked-list and return it.

## LeetCode URL: https://leetcode.com/problems/merge-k-sorted-lists/description/

```python
class LinkedList:

    '''
    Title: Creating a linked list
    Below code creates a linked list. we have created two methods one is linked list and another is Node
    the we are manually creating five elements.
    Then we are printing the elements using printlist method created in class.
    '''

    def __init__(self,val):
        self.head = val
    
    def _printList(self):
        temp = self.head
        while temp:
            print(temp.node,end=" ")
            temp = temp.next

class Node:

    def __init__(self,val):
        self.node = val
        self.next = None


if __name__ == '__main__':

    first = Node(1)
    second = Node(2)
    third = Node(3)
    fourth = Node(4)
    fifth = Node(5)
    sixth = Node(6)
    seven = Node(7)
    eight = Node(8)
    nine = Node(9)
    tenth = Node(10)
    eleventh = Node(11)
    twelth = Node(12)
    thirteenth = Node(13)

    first.next = third
    third.next = fifth
    fifth.next = eight

    second.next = fourth
    fourth.next = sixth
    sixth.next = seven
    seven.next = nine

    tenth.next = eleventh
    eleventh.next = twelth
    twelth.next = thirteenth


    # Sorted List 1
    ll = LinkedList(first)

    # Sorted List 2
    mm = LinkedList(second)

    # Sorted List 3
    nn = LinkedList(tenth)

    
mm._printList()
```

```python
def mergeklists(lists):

    '''
    Merging K sorted lists will use similar concepts that are there in Merger sort.
    For merging two sorted lists we used a dummy vairable and we pointed the next of dummy to lowest of two lists
    and at the end we retrned dummy. next which is the head of our new sorted list.
    But now as we are talking about k linked lists we will just take two elements of lists at a time, merge those two
    using above method and then append it back to the initial list. Now we will repeat the process till
    len of list is only 1 , so it will mean we have sorted all the k lists and appended them to initial list.
    
    Args:
    a. list[list]

    Output:
    list
    '''
    if not lists or len(lists) ==0:
        return None                        

    def mergetwolists(l1,l2):
        dummy = Node(float('inf'))
        temp = dummy
        while l1 and l2:

            if l1.node <= l2.node:
                temp.next = l1
                l1 = l1.next
            else:
                temp.next = l2
                l2 = l2.next
            temp = temp.next

        if l1:
            temp.next = l1
        
        if l2:
            temp.next = l2

        return dummy.next


    while len(lists) > 1:
        mergedLists = []

        for i in range(0,len(lists),2):
            l1 = lists[i]
            l2 = lists[i+1] if (i+1) < len(lists) else None
            mergedLists.append(mergetwolists(l1,l2))
        lists = mergedLists


    return lists[0]  

lists = [ll.head,mm.head,nn.head]

temp = mergeklists(lists)

while temp:
    print(temp.node, end=" ")
    temp = temp.next
```
